### Analysis

In this analysis of Pymaceuticals, Inc.'s study on potential treatments for SCC, I dug into a dataset encompassing 249 mice subjected to various drug regimens. My primary focus was on comparing the performance of Pymaceuticals' drug of interest, Capomulin, against alternative treatments. Initial steps involved data preparation, cleaning, and exploration, including the identification and removal of a mouse with duplicate time points. I generated summary statistics tables, revealing key insights into the efficacy of different drug regimens. My visualization efforts included bar charts and pie charts illustrating the distribution of mice across treatments and gender, contributing to a comprehensive understanding of the study population. I looked at the final tumor volumes for select treatment regimens, highlighting potential outliers with box plots. Additionally, I explored individual mouse responses to Capomulin treatment, showcasing a line plot for tumor volume over time and a scatter plot demonstrating the correlation between mouse weight and average tumor volume. Rigorous statistical analysis, including correlation and linear regression, added very valuable insight to our Data Analysis. This in depth approach equips Pymaceuticals, Inc. with a thorough data report on the newest potential anti-cancer drug, Capomulin.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
merged_df.head()
```

```python
# Checking the number of mice.
num_mice = merged_df['Mouse ID'].nunique()
print(f"Number of unique mice: {num_mice}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merged_df[merged_df.duplicated(['Mouse ID', 'Timepoint'])]
print(f"Duplicate Mouse ID and Timepoints: {duplicate_mice['Mouse ID'].unique()}")
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mouse_data = merged_df[merged_df['Mouse ID'] == 'g989']
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = merged_df[merged_df['Mouse ID'] != 'g989']
cleaned_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_num_mice = cleaned_df['Mouse ID'].nunique()
print(f"Number of unique mice in the clean DataFrame: {clean_num_mice}")
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = cleaned_df.groupby('Mouse ID')['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume_df = pd.merge(last_timepoint, cleaned_df, on=['Mouse ID', 'Timepoint'], how='inner')
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol = final_tumor_volume_df.loc[final_tumor_volume_df['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol_data.append(tumor_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = tumor_vol.loc[(tumor_vol < lower_bound) | (tumor_vol > upper_bound)]
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol_data, labels=treatments, flierprops=dict(markerfacecolor='r', marker='D'))
plt.title('Final Tumor Volume for Treatment Regimens')
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_mouse = cleaned_df.loc[(cleaned_df['Drug Regimen'] == 'Capomulin') & (cleaned_df['Mouse ID'] == 's185')]
plt.plot(capomulin_mouse['Timepoint'], capomulin_mouse['Tumor Volume (mm3)'])
plt.title('Capomulin Treatment - Tumor Volume vs Timepoint')
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen.
# This was terribly confusing. I could not figure out for the life of me why the 'mean()' function was causing so many issues.
# I replaced .mean() with .agg() and specified the columns for which we want to calculate the mean.

capomulin_avg_vol = cleaned_df[cleaned_df['Drug Regimen'] == 'Capomulin'].groupby('Mouse ID').agg({
    'Weight (g)': 'mean',
    'Tumor Volume (mm3)': 'mean'})
plt.scatter(capomulin_avg_vol['Weight (g)'], capomulin_avg_vol['Tumor Volume (mm3)'])
plt.title('Capomulin Treatment - Mouse Weight vs Avg Tumor Volume')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(capomulin_avg_vol['Weight (g)'], capomulin_avg_vol['Tumor Volume (mm3)'])
print(f"Correlation between mouse weight and average observed tumor volume: {correlation[0]:.4f}") #I forgot how many places we needed to round to, 4 seems good.

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(capomulin_avg_vol['Weight (g)'], capomulin_avg_vol['Tumor Volume (mm3)'])
regress_values = capomulin_avg_vol['Weight (g)'] * slope + intercept
plt.scatter(capomulin_avg_vol['Weight (g)'], capomulin_avg_vol['Tumor Volume (mm3)'])
plt.plot(capomulin_avg_vol['Weight (g)'], regress_values, color='red')
plt.title('Capomulin Treatment - Mouse Weight vs Avg Tumor Volume with Linear Regression')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
